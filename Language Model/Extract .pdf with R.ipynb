Library to extract text from .pdf

```r
library(pdftools)
```

root path:

```r
path <- './data/Kinderbücher/einfachvorlesen/'
```

Get all dircetories in the root path, append `/`.

```r
dirs <- list.dirs(path = "./data/Kinderbücher", full.names = TRUE, recursive = FALSE)
m <- length(dirs)

for(i in 1:m){
    dirs[i] <- paste(dirs[i], '/', sep = '')
}
```

For all directories, get the contained .pdf files. For every file (concat the current directory path), read the .pdf and extract it to a .txt.

```r
for(dir in dirs){
    # get .pdf files from current dir
    files <- list.files(path = dir, pattern = 'pdf$')
    n <- length(files)
    
    # concat the current dir-path and the filenames
    for(i in 1:n){
        files[i] <- paste(dir, files[i], sep = '')
    }
    
    for(i in 1:n){
        # read pdf file
        tmp <- pdf_text(files[i])
        # copy the old path & filename, exchange .pdf for .txt
        to <- gsub('.pdf', '.txt', files[i])
        # write pdf file .txt file
        cat(tmp, file = to)
    }
}
```
